```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
ns = [2 , 5 , 10 , 20 , 100 , 1000]
fig, ax = plt.subplots(2 , 3 )
fig.set_figwidth(20)
plt.subplots_adjust(bottom = -1)
ax = ax.flatten()
for plot_no,n in enumerate(ns):
    toss = np.random.binomial(n  , p = 0.5 , size = 100)
    ax[plot_no].hist(toss , color= 'green', edgecolor='black', linewidth=1.2)
    ax[plot_no].set_xlabel('the number of heads')
    ax[plot_no].set_ylabel('the number of samples')
    ax[plot_no].set_xlim([0 , n])
    ax[plot_no].set_title('n=%d'%n)
plt.show()
```
